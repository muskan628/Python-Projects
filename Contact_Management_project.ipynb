```python
import csv
import os
csv_file="contact.csv"

def load_contacts():
  contacts = {}
  if os.path.exists(csv_file):
    with open(csv_file, 'r') as file:
      reader = csv.reader(file)
      next(reader) # skip the first row
      for row in reader:
        contact_id = int(row[0])
        contacts[contact_id] = {"name": row[1], "phone": row[2], "email": row[3]}
  return contacts

def add_contacts(contacts):
  name=input("Enter the name:")
  phone=input("Enter the phone number:")
  email=input("Enter the email:")
  contact_id=max(contacts.keys())+1 if contacts else 1
  contacts[contact_id] = {"name":name,"phone": phone,"email":email}
  print(f"contact {name} is added successfully with contact id {contact_id}")

  save_contacts(contacts)
  view_contacts(contacts)

def delete_contacts(contacts):
  contact_id=int(input("Enter the contact id:"))
  if contact_id in contacts:
    del contacts[contact_id]
    print("contact deleted successfully")
  else:
    print("contact not found")

  save_contacts(contacts)
  view_contacts(contacts)

def save_contacts(contacts):
  with open(csv_file, "w") as file:
    writer=csv.writer(file)
    writer.writerow(["id","name","phone","email"])
    for contact_id,contact_info in contacts.items():
      writer.writerow([contact_id,contact_info["name"],contact_info["phone"],contact_info["email"]])

def view_contacts(contacts):
  if not contacts:
    print("No contacts found")
  else:
    print("\nContacts:")
    for (contact_id, contact_info) in contacts.items():
      print(f'ID: {contact_id}, Name: {contact_info["name"]}, Phone: {contact_info["phone"]}, Email: {contact_info["email"]}')

def main():
  contacts = load_contacts()  # return all the contacts that are stored in csv file.

  while True:
    print("Contact Management System")
    print("1.Add Contact")
    print("2.Delete Contact")
    print("3.View Contacts")
    print("4.Exit")

    choice=input("Enter your choice:")

    print()
    if choice=="1":
      add_contacts(contacts)
    elif choice=="2":
      delete_contacts(contacts)
    elif choice=="3":
      view_contacts(contacts)
    elif choice=="4":
      print("Goodbye and have a great day")
      break
    else:
      print("Invalid choice.please try again")

main()
```
