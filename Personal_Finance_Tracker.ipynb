```python
# Class to represent a single transaction
class Transaction:
    def __init__(self, transaction_id, date, t_type, amount, category):
        self.id = transaction_id
        self.date = date
        self.t_type = t_type
        self.amount = amount
        self.category = category

    def __str__(self):
        return f"ID: {self.id}, Date: {self.date}, Type: {self.t_type}, Amount: {self.amount}, Category: {self.category}"


# Class to manage all transactions
class FinanceManager:
    def __init__(self):
        self.transactions = []

    def add_transaction(self, transaction):
        self.transactions.append(transaction)
        print(f"Added transaction with ID {transaction.id}.")
        print()

    def remove_transaction(self, transaction_id):
        transaction_to_remove = next((t for t in self.transactions if t.id == transaction_id), None)
        if transaction_to_remove:
            self.transactions.remove(transaction_to_remove)
            print(f"Removed transaction with ID {transaction_id}.")
            print()
        else:
            print(f"Transaction with ID {transaction_id} not found.")
            print()

    def list_transactions(self):
        if not self.transactions:
            print("No transactions available.")
            print()
        else:
            for transaction in self.transactions:
                print(transaction.__str__())
            print()

    def summarize_totals(self):
        income = sum(t.amount for t in self.transactions if t.t_type == "income")   ## [1,  income, 10 ] [ 2, expense, 20]  [ 3, income , 20]
        expense = sum(t.amount for t in self.transactions if t.t_type == "expense")
        print(f"Total Income: {income:.2f}, Total Expenses: {expense:.2f}")
        print()


def menu():
  manager = FinanceManager()
  while True:
    print("Personal Finance Tracker Menu: \n1. Add a new transaction \n2. Remove a transaction \n3. List all transactions \n4. Summarize income and expenses \n5. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
      transaction_id=int(input("Enter transaction ID: "))

      date=input("Enter transaction date (yyyy-mm-dd): ")

      t_type=input("Enter transaction type (Income/Expense): ").strip().lower()
      if t_type not in ["income", "expense"]:
                print("Invalid choice, setting to 'income' by default.")
                t_type = "income"

      amount=float(input("Enter transaction amount: "))

      category= input("transaction category (Salary, Groceries, Entertainment, Bills, Other): ").strip().capitalize()
      if category not in ["Salary", "Groceries", "Entertainment", "Bills", "Other"]:
                print("Invalid category, setting to 'Other' by default.")
                category = "Other"


      new_transaction = Transaction(transaction_id, date, t_type, amount, category)
      manager.add_transaction(new_transaction)

    elif choice == "2":
      transaction_id = int(input("Enter the transaction ID to remove: "))
      manager.remove_transaction(transaction_id)

    elif choice == "3":
      manager.list_transactions()

    elif choice == "4":
      manager.summarize_totals()

    elif choice == "5":
      print("Exiting the Personal Finance Tracker. Goodbye!")
      break

    else:
      print("Invalid choice. Please enter a number from 1 to 5.")


menu()
```
