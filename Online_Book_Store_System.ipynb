```python
# Initialize the BookStore
BookStore = [
    ("The Great Gatsby", "F. Scott Fitzgerald", 1925, "Fiction"),
    ("To Kill a Mockingbird", "Harper Lee", 1960, "Fiction"),
    ("1984", "George Orwell", 1949, "Dystopian"),
    ("Pride and Prejudice", "Jane Austen", 1813, "Romance"),
]

# Display all books at the start
print("\nListing all books:")
if not BookStore:
    print("The BookStore is empty.")
else:
    print("BookStore Collection:")
    for title, author, year, genre in BookStore:
        print(f"Title: {title}, Author: {author}, Year: {year}, Genre: {genre}")

# Adding a new book to the BookStore
title = input("\nEnter the title of the book you want to add: ")
author = input("Enter the author of the book: ")
year = int(input("Enter the publication year: "))
genre = input("Enter the genre of the book: ")

# Create a set of existing titles for easy lookup
existing_titles = {book[0] for book in BookStore}

# Check if the title is already in the BookStore
if title in existing_titles:
    print(f"'{title}' already exists in the BookStore.")
else:
    new_book = (title, author, year, genre)
    BookStore.append(new_book)
    print(f"Added '{title}' to the BookStore.")

# Display all books after adding a new book
print("\nListing all books:")
if not BookStore:
    print("The BookStore is empty.")
else:
    print("BookStore Collection:")
    for title, author, year, genre in BookStore:
        print(f"Title: {title}, Author: {author}, Year: {year}, Genre: {genre}")

# Removing a book from the BookStore
title_to_remove = input("\nEnter the title of the book you want to remove: ")

for book in BookStore:
    if book[0] == title_to_remove:
        BookStore.remove(book)
        print(f"Removed '{title_to_remove}' from the BookStore.")
        break

else:
    print(f"'{title_to_remove}' not found in the BookStore.")


# Display all books after removing a book
print("\nListing all books:")
if not BookStore:
    print("The BookStore is empty.")
else:
    print("BookStore Collection:")
    for title, author, year, genre in BookStore:
        print(f"Title: {title}, Author: {author}, Year: {year}, Genre: {genre}")

# Displaying unique genres in the BookStore
genres = {book[3] for book in BookStore}  # Extract genres from each book
print(f"\nUnique genres in the BookStore: {genres}")

# Updating book information
title_to_update = input("\nEnter the title of the book you want to update: ")
detail_to_update = input("Enter the detail you want to update (title, author, year, genre): ").lower()
new_value = input(f"Enter the new value for {detail_to_update}: ")

for index, book in enumerate(BookStore):
    if book[0] == title_to_update:
        book_details = list(book)
        if detail_to_update == "title":
            book_details[0] = new_value
        elif detail_to_update == "author":
            book_details[1] = new_value
        elif detail_to_update == "year":
            book_details[2] = int(new_value)
        elif detail_to_update == "genre":
            book_details[3] = new_value
        else:
            print("Invalid detail to update.")
            break
        BookStore[index] = tuple(book_details)
        print(f"Updated '{title_to_update}' with new {detail_to_update}: {new_value}")
        break
else:
    print(f"'{title_to_update}' not found in the BookStore.")

# Display all books after updating a book
print("\nListing all books:")
if not BookStore:
    print("The BookStore is empty.")
else:
    print("BookStore Collection:")
    for title, author, year, genre in BookStore:
        print(f"Title: {title}, Author: {author}, Year: {year}, Genre: {genre}")
```
